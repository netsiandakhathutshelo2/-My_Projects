```python
#clean the data
clean_data = df.dropna()
clean_data.tail()
```

```python
#the cleaned data
clean_data.shape
```

```python
#Find College Major with Highest Starting Salaries

max_salary = clean_data[clean_data["Starting Median Salary"] == clean_data["Starting Median Salary"].max()]
max_salary = max_salary["Undergraduate Major"]
f"College Major with Highest Starting Salaries: {max_salary.to_string()} "
```

```python
#Find College Major with Highest Starting Salaries

minimum_salary = clean_data[clean_data["Starting Median Salary"] == clean_data["Starting Median Salary"].min()]
minimum_salary =  minimum_salary["Undergraduate Major"]
f"College Major with Highest Starting Salaries: {minimum_salary.to_string()} "
```
